```python
import requests
import time
import os

def downfile(file, url):      #
    print("Start Downloading：", url)
    try:
        r = requests.get(url, stream=True,timeout=30)
        
        with open(file, 'wb') as fd:
            for chunk in r.iter_content():    # iter_content：一块一块的遍历要下载的内容
                fd.write(chunk)
    except Exception as e:   
        print("Download failed.", e)
```

```python
urls = []
with open("Maltese.txt", "r+") as f:
    for line in f.readlines():
        urls.append(line.rstrip())
i=0
for url in urls:
    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.54 Safari/537.36'}
    PWD = 'E:\Workshop\image'
    name = url.split('/')[-1]
  
    i += 1 
    
    filename =  PWD +str(i) + '.jpg'
   
    if os.path.isfile(filename):  
        print("Documents exist：", filename)
        continue
    try:
        r = requests.get(url, headers=headers, stream=True,timeout=30)
    except Exception as e:   
        print("The request failed.", e)
        continue
   
    downfile(filename, url)
print('--------------------------------')
print('over!')
```
